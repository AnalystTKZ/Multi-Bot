```python
from pathlib import Path
import shutil
import os

INPUT_BASE = Path(CODE_PATH)
WORK_BASE  = Path("/kaggle/working/Multi-Bot/trading-system")

print("рџ”„ HARD SYNC INPUT в†’ WORK (mirror mode)")

WORK_BASE.mkdir(parents=True, exist_ok=True)

# 1. Delete anything in WORK that no longer exists in INPUT
for item in WORK_BASE.rglob("*"):
    rel = item.relative_to(WORK_BASE)
    src_equiv = INPUT_BASE / rel

    if not src_equiv.exists():
        if item.is_dir():
            shutil.rmtree(item)
        else:
            item.unlink()

# 2. Copy fresh input в†’ work (overwrite everything)
for item in INPUT_BASE.rglob("*"):
    rel = item.relative_to(INPUT_BASE)
    target = WORK_BASE / rel

    if item.is_dir():
        target.mkdir(parents=True, exist_ok=True)
    else:
        target.parent.mkdir(parents=True, exist_ok=True)
        shutil.copy2(item, target)

print("вњ… Mirror sync complete (WORK == INPUT)")
```

```python
"""
kaggle_train.py вЂ” MINIMAL Kaggle retraining entrypoint

вњ” Uses existing processed_data
вњ” Skips pipeline entirely
вњ” Runs ONLY retrain_incremental
вњ” Exports weights
"""

from __future__ import annotations
import os
import sys
import shutil
import subprocess
from pathlib import Path

from pathlib import Path
import os
import sys

# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# 1. PATH SETUP
# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

INPUT_BASE = Path("/kaggle/input/datasets/tysonsiwela/multi-bot-system/trading-system")
WORK_BASE  = Path("/kaggle/working/Multi-Bot/trading-system")

# Ensure working directory exists
os.makedirs(WORK_BASE, exist_ok=True)

# вњ… FIX: use WORK_BASE instead of undefined HERE
os.chdir(WORK_BASE)
sys.path.insert(0, str(WORK_BASE))

from env_config import get_env, ensure_output_dirs

env = get_env()
ensure_output_dirs(env)

print("\n=== KAGGLE RETRAIN START ===")
print(f"Base      : {env['base']}")
print(f"Processed : {env['processed']}")
print(f"Weights   : {env['weights']}")
print(f"Output    : {env['output']}")

# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# 2. VERIFY REQUIRED DATA
# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

required_data = [
    env["processed"] / "histdata" / "EURUSD_5M.parquet",
    env["processed"] / "histdata" / "XAUUSD_1D.parquet",
]

missing = [p for p in required_data if not p.exists()]
if missing:
    raise FileNotFoundError(f"Missing processed data: {missing}")

print("\nвњ“ Processed data verified")

# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# 3. RUN RETRAIN SCRIPT ONLY
# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

retrain_script = env["base"] / "kaggle_train.py"

if not retrain_script.exists():
    raise FileNotFoundError(f"Missing retrain script: {retrain_script}")

print("\n=== RUNNING RETRAIN ===\n")

result = subprocess.run(
    [sys.executable, str(retrain_script)],
    cwd=str(env["base"]),
    env={
        **os.environ,
        "PYTHONPATH": f"{env['base']}:{env['base'] / 'trading-engine'}",
    },
)

if result.returncode != 0:
    raise RuntimeError("Retraining FAILED")

print("\n=== RETRAIN COMPLETE ===")

# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# 4. EXPORT WEIGHTS
# в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

output_weights = env["output"] / "trained_weights"

if env["weights"].exists():
    shutil.copytree(env["weights"], output_weights, dirs_exist_ok=True)
    print(f"\nWeights exported в†’ {output_weights}")
else:
    print("\nNo weights found")
```
