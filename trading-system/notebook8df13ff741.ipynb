```python
from pathlib import Path
import shutil
import os

INPUT_BASE = Path(CODE_PATH)
WORK_BASE  = Path("/kaggle/working/Multi-Bot/trading-system")

print("🔄 HARD SYNC INPUT → WORK (mirror mode)")

WORK_BASE.mkdir(parents=True, exist_ok=True)

# 1. Delete anything in WORK that no longer exists in INPUT
for item in WORK_BASE.rglob("*"):
    rel = item.relative_to(WORK_BASE)
    src_equiv = INPUT_BASE / rel

    if not src_equiv.exists():
        if item.is_dir():
            shutil.rmtree(item)
        else:
            item.unlink()

# 2. Copy fresh input → work (overwrite everything)
for item in INPUT_BASE.rglob("*"):
    rel = item.relative_to(INPUT_BASE)
    target = WORK_BASE / rel

    if item.is_dir():
        target.mkdir(parents=True, exist_ok=True)
    else:
        target.parent.mkdir(parents=True, exist_ok=True)
        shutil.copy2(item, target)

print("✅ Mirror sync complete (WORK == INPUT)")
```

```python
from __future__ import annotations
import os
import sys
import shutil
import subprocess
from pathlib import Path

# ─────────────────────────────────────────────────────────────
# 1. PATH SETUP
# ─────────────────────────────────────────────────────────────

WORK_BASE = Path("/kaggle/working/Multi-Bot/trading-system")
WORK_BASE.mkdir(parents=True, exist_ok=True)
os.chdir(WORK_BASE)
sys.path.insert(0, str(WORK_BASE))

# DATA_PATH is set in cell 0 — pass it to env_config so it finds the data dataset
# (training_data/ + processed_data/) even at non-standard Kaggle mount paths
os.environ["KAGGLE_DATA_PATH"] = str(DATA_PATH)

from env_config import get_env, ensure_output_dirs

env = get_env()
ensure_output_dirs(env)

print("\n=== KAGGLE RETRAIN START ===")
print(f"Base      : {env['base']}")
print(f"Data      : {env['data']}")
print(f"Processed : {env['processed']}")
print(f"Weights   : {env['weights']}")
print(f"Output    : {env['output']}")

# ─────────────────────────────────────────────────────────────
# 2. RESTORE WEIGHTS + LOGS FROM PREVIOUS DATASET VERSION
# ─────────────────────────────────────────────────────────────
# If the data dataset has weights/logs from a previous training run, restore
# them into the working copy so warm-start retraining picks them up.

data_weights = Path(DATA_PATH) / "weights"
data_logs    = Path(DATA_PATH) / "logs"

if data_weights.exists() and not env["weights"].exists():
    shutil.copytree(str(data_weights), str(env["weights"]))
    print(f"\n✓ Restored weights from dataset: {data_weights}")
elif data_weights.exists():
    # Merge: dataset weights fill in missing files, working weights win on conflict
    for f in data_weights.rglob("*"):
        if f.is_file():
            dst = env["weights"] / f.relative_to(data_weights)
            if not dst.exists():
                dst.parent.mkdir(parents=True, exist_ok=True)
                shutil.copy2(f, dst)
    print(f"✓ Merged weights from dataset (warm start)")

if data_logs.exists():
    for f in data_logs.rglob("*"):
        if f.is_file():
            dst = env["engine"] / "logs" / f.relative_to(data_logs)
            if not dst.exists():
                dst.parent.mkdir(parents=True, exist_ok=True)
                shutil.copy2(f, dst)
    print(f"✓ Restored logs from dataset (journal for RL retraining)")

# ─────────────────────────────────────────────────────────────
# 3. VERIFY REQUIRED DATA
# ─────────────────────────────────────────────────────────────

required = [
    env["processed"] / "histdata" / "EURUSD_5M.parquet",
    env["processed"] / "histdata" / "XAUUSD_1D.parquet",
]
missing = [p for p in required if not p.exists()]
if missing:
    raise FileNotFoundError(f"Missing histdata parquets: {missing}\nEnsure DATA_PATH dataset contains processed_data/histdata/")

print("\n✓ Histdata parquets verified")

# ─────────────────────────────────────────────────────────────
# 4. RUN FULL PIPELINE (kaggle_train.py)
# ─────────────────────────────────────────────────────────────

retrain_script = env["base"] / "kaggle_train.py"
if not retrain_script.exists():
    raise FileNotFoundError(f"kaggle_train.py not found: {retrain_script}")

print("\n=== RUNNING PIPELINE ===\n")

result = subprocess.run(
    [sys.executable, str(retrain_script)],
    cwd=str(env["base"]),
    env={
        **os.environ,
        "KAGGLE_DATA_PATH": str(DATA_PATH),
        "PYTHONPATH": f"{env['base']}:{env['base'] / 'trading-engine'}",
    },
)

if result.returncode != 0:
    raise RuntimeError("Pipeline FAILED — check logs above")

print("\n=== PIPELINE COMPLETE ===")

# ─────────────────────────────────────────────────────────────
# 5. SAVE OUTPUTS FOR DATASET UPDATE
# ─────────────────────────────────────────────────────────────
# Copy weights + logs to /kaggle/working/outputs/processed_data/
# so you can re-upload them as the next version of the data dataset.

save_root = env["output"] / "outputs" / "processed_data"
save_root.mkdir(parents=True, exist_ok=True)

if env["weights"].exists():
    shutil.copytree(str(env["weights"]), str(save_root / "weights"), dirs_exist_ok=True)
    print(f"\nWeights saved → {save_root / 'weights'}")

engine_logs = env["engine"] / "logs"
if engine_logs.exists():
    shutil.copytree(str(engine_logs), str(save_root / "logs"), dirs_exist_ok=True)
    print(f"Logs saved    → {save_root / 'logs'}")

print(f"\nDataset update bundle: {save_root}")
print("Download /kaggle/working/outputs/ and re-upload as the data dataset version.")
```
